```python
import matplotlib.pyplot as plt

# Sort the clothing data in descending order of count and take the top 10 items
sorted_data = sorted(data, key=lambda x: x[1], reverse=True)[:20]

# Extract the item names and counts from the sorted data
items = [item[0] for item in sorted_data]
counts = [item[1] for item in sorted_data]

# Create a bar graph of the top 10 items and their counts
plt.bar(items, counts)
plt.xticks(rotation=90)
plt.xlabel('Clothing Item')
plt.ylabel('Count')
plt.title('Top 10 Clothing Items by Count')
plt.show()
```

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# extract all words from pant_items dataframe
all_words = []
for desc in pant_items['Cloth_Desc']:
    words = desc.split()
    all_words.extend(words)

# count frequency of each word
word_counts = Counter(all_words)

# create a dataframe of the 10 most common words
common_words = pd.DataFrame(word_counts.most_common(10), columns=['Word', 'Count'])

# create a bar chart of the top 10 words and their counts
plt.bar(common_words['Word'], common_words['Count'])
plt.title('Top 10 Words in Pant Items')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```
