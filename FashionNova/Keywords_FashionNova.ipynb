```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
file_path = r"C:\Users\Braydon\Desktop\Shein_Data\Pivot_Data\FashionNova\FashionNova_Data_output.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
df
```

```python
cloth_word_lst = []

for item in df['Cloth_Desc']:
    item = item.split()
    for thing in item:
        cloth_word_lst.append(thing)
```

```python
cloth_word_count = Counter(cloth_word_lst)
```

```python
cloth_word_count
```

```python
data = cloth_word_count.most_common()
```
